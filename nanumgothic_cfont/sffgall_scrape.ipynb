```python
import requests
from bs4 import BeautifulSoup
from collections import Counter
import re

GALLERY_ID = 'sff'
BASE_URL = f'https://gall.dcinside.com/mgallery/board/lists?id={GALLERY_ID}'

headers = {
    'User-Agent': 'Mozilla/5.0',
}

def get_titles(page=1):
    url = f'{BASE_URL}&page={page}'
    res = requests.get(url, headers=headers)
    soup = BeautifulSoup(res.text, 'html.parser')
    
    titles = soup.select('td.gall_tit.ub-word > a:nth-child(1)')
    return [title.text.strip() for title in titles if title.text.strip()]

def count_characters(titles):
    all_text = ''.join(titles)
    filtered = re.sub(r'[^ㄱ-힣]', '', all_text)
    return Counter(filtered)

if __name__ == "__main__":
    total_titles = []
    
    for page in range(1, 101):
        titles = get_titles(page)
        total_titles.extend(titles)

    char_counts = count_characters(total_titles)
```

```python
characters = list(char_counts.keys())
characters.sort()
```

```python
from itertools import groupby
from operator import itemgetter

codepoints = sorted([ord(c) for c in characters])

def group_code_ranges(codepoints):
    ranges = []
    for k, g in groupby(enumerate(codepoints), lambda x: x[1] - x[0]):
        group = list(map(itemgetter(1), g))
        ranges.append((group[0], group[-1]))
    return ranges

ranges = group_code_ranges(codepoints)

for start, end in ranges:
    if start == end:
        print(f"{start}", end=',')
    else:
        print(f"{start}-{end}", end=',')
print('\b')
```
